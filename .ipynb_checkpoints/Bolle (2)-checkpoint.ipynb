```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.signal import find_peaks

def bubble(x, amp, cen1, cen2, offset, wid1, wid2):
    return -amp*( np.arctan((x-cen1)/wid1) - np.arctan((x-cen2)/wid2) -2) + offset

def gauss(x, amp, cen1, wid1, offset):
    return -amp*(np.exp(-(x-cen1)**2/2/wid1**2)) + offset

def bubbleshoulder(x, amp, cen1, offset, wid1):
    return -amp*( np.arctan((x-cen1)/wid1)) + offset

def bubble_growA(x, tau, sat, amp):
    y = (amp*np.exp(x/tau)-1)
    mask = (y>=sat)
    y[mask] = sat
    return y

def bubble_grow(x, tau, sat):
    y = 0.002*(np.exp(x/tau)-1)
    mask = (y>=sat)
    y[mask] = sat
    return y
```

```python
#f = [r"/home/gabriele/NAS542_dataBEC2/2022/08/12/seq_89.hdf"]
f = [r"/home/gabriele/NAS542_dataBEC2/2022/08/12/seq_89.hdf",r"/home/gabriele/NAS542_dataBEC2/2022/08/17/seq_2.hdf",r"/home/gabriele/NAS542_dataBEC2/2022/08/18/seq_8.hdf"]
seqs = [[5,6,7,8],[8,9,10],[10,14]]
#seqs = [[5]]
guess_grow = np.array([[2,10,50,50],[50,50,50],[2,10]])
detuning = [[0.473, 0.439, 0.270, 0.338], [0.169, 0.236, 0.304],[0.473, 0.371]]

grow_time = []
half_time = []
min_bubble = []
grow_amp = []
slope_bubble = []
fin_bubble = []
    
for fs in [0,1,2]: 
    df12_ = pd.read_hdf(f[fs])
    #print(np.unique(df12_['sequence_index']))
    
    for ei, ii in enumerate(seqs[fs]):
        df12 = df12_[(df12_['sequence_index'].isin([ii]))]
        df12 = df12[(df12[('od_remove_thpart', 'N_bec')]>2e5)]
        df12 = df12[(df12[('od_remove_thpart', 'N_bec')]<2e6)]
        y_axis = 'uW_pulse'
        df12 = df12.sort_values(y_axis)
        #print(df12.shape)

        m1 = df12[('od_remove_thpart', 'm1_1d')]
        m2 = df12[('od_remove_thpart', 'm2_1d')]
        rep = df12[('Repetition')]
        time = df12[('uW_pulse')]

        shape = (800, )
        check = type(np.ndarray(1))
        good = np.array([type(p) == check for p in m1])
        goodgood = [p.shape == (800,) for p in m1[good]]
        good[good] = goodgood

        c= 400
        w= 180
        roi = np.s_[:, c-w:c+w]
        M1 = m1[good]
        M2 = m2[good]
        D = M1+M2*1.3
        M = (M2*1.3-M1)/D
        D = np.vstack(D)[roi]
        M = np.vstack(M)[roi]
        mask_D = ((np.sum(D, axis = 1)>1.1e5) & (np.sum(D, axis = 1)<1.8e5))
        time = time[good]
        D = D[mask_D]
        M = M[mask_D]
        time = np.array(time)[mask_D]
        print(M.shape)
        
        fig, ax = plt.subplots(figsize = (10, 5), ncols=2)
        ax[0].pcolormesh(M, vmin = -1, vmax = +1, cmap = 'RdBu')
        ax[1].pcolormesh(D)
        
        b_check=20
        s_size=30
        b_check_roi = np.s_[:, w-b_check:w+b_check]
        Mb=np.mean(M[b_check_roi],axis=1)
        
        xx=np.arange(2*w)
        bsize=[]
        bslope=[]
        bsizeADV=[]
        bslopeADV=[]
        times=np.unique(time)
        #print(times)
        #plt.plot(xx,bubble(xx,0.4,110,290,-0.1,10,10))
        #fig, ax = plt.subplots(nrows = 3)
        #ax[0].plot(np.sum(D, axis = 1))
        #ax[1].plot(df12[('od_remove_thpart', 'N_bec')].values)
        #ax[2].plot(df12[('od_remove_thpart', 'radius')].values)        


        for i in np.arange(len(M)):
            init_vals = [0.3, 140, 160, 0 , 10, 10]  # for [amp, cen, wid]
            #plt.plot(xx,M[i])
            #plt.ylim(-1,1)
            if Mb[i]<0.2:

                try:
                    best_vals, covar = curve_fit(bubble, xx, M[i], p0=init_vals)
                    init_BS_left= [best_vals[0],best_vals[1],best_vals[3],best_vals[4]]
                    best_BS_left, covar_BS_left = curve_fit(bubbleshoulder, xx[int(round(best_vals[1]))-s_size:int(round(best_vals[1]))+s_size], M[i][int(round(best_vals[1]))-s_size:int(round(best_vals[1]))+s_size], p0=init_BS_left)
                    init_BS_right= [-best_vals[0],best_vals[2],best_vals[3],best_vals[5]]
                    best_BS_right, covar_BS_right = curve_fit(bubbleshoulder,xx[int(round(best_vals[2]))-s_size:int(round(best_vals[2]))+s_size], M[i][int(round(best_vals[2]))-s_size:int(round(best_vals[2]))+s_size], p0=init_BS_right)
                    xl=xx[int(round(best_vals[1]))-s_size:int(round(best_vals[1]))+s_size]
                    xr=xx[int(round(best_vals[2]))-s_size:int(round(best_vals[2]))+s_size]
                    #plt.plot(xx,bubble(xx,*best_vals))
                    #plt.plot(xl,bubbleshoulder(xl,*best_BS_left))
                    #plt.plot(xr,bubbleshoulder(xr,*best_BS_right))
                    bsize.append(best_vals[2]-best_vals[1])
                    bsizeADV.append(best_BS_right[1]-best_BS_left[1])
                    bslope.append(best_vals[4]/2+best_vals[5]/2)
                    bslopeADV.append(best_BS_right[3]/2+best_BS_left[3]/2)
                    #print("arctan")
                except:
                    best_GS, covar_GS = curve_fit(gauss, xx, M[i], p0=[.4,150,2,.5])
                    #plt.plot(xx,gauss(xx,*best_GS))
                    bsize.append(best_GS[2]*2.355)
                    bsizeADV.append(best_GS[2]*2.355)
                    bslope.append(0)
                    bslopeADV.append(0)
                    #print("gauss")
                    
                #print(Mb[i])
                #print(np.std(M[b_check_roi],axis=1)[i])
  
            else:
                best_vals=[0,0,0,0,0,0]
                best_BS_left=[0,0,0,0]
                best_BS_right=[0,0,0,0]
                bsize.append(0)
                bslope.append(0)
                bsizeADV.append(0)
                bslopeADV.append(0)
            #plt.show()
            #print(best_vals)

        bsize=np.array(bsize)
        bsizeADV=np.array(bsizeADV)
        bmin=np.min(bsize[np.where(bsize>0)])
        bslope=np.array(bslope)  
        bslopeADV=np.array(bslopeADV) 
        min_bubble.append(bmin)

        #print(bmin)
        #print(time.shape)
        #print(len(M))
        bubble_prop=[]
        bubble_mean=[]
        bubble_slope=[]
        bubble_meanADV=[]
        bubble_slopeADV=[]
        bubble_slopeADVerr=[]
        bubble_mean_err=[]
        for i in times:
            sam=np.count_nonzero(time == i)
            #print(np.where(time==i))
            selection=bsize[np.where(time==i)]
            selection2=bslope[np.where(time==i)]
            selectionADV=bsizeADV[np.where(time==i)]
            selection2ADV=bslopeADV[np.where(time==i)]
            bubble_prop.append(np.count_nonzero(selection > 0)/sam)
            bubble_slope.append(np.mean(selection2[np.where(selection> 0)]))
            bubble_meanADV.append(np.mean(selectionADV[np.where(selectionADV> 0)]))
            bubble_slopeADV.append(np.mean(selection2ADV[np.where(selectionADV> 0)]))
            bubble_slopeADVerr.append(np.std(selection2ADV[np.where(selectionADV> 0)]))
            bubble_mean.append(np.mean(selection[np.where(selection> 0)]))
            bubble_mean_err.append(np.std(selection[np.where(selection> 0)]))
        #print(bubble_mean)
        #print(bubble_prop)
        
        
        fig, ax = plt.subplots(nrows = 4)
        ax[0].plot(times,bubble_mean)
        ax[0].plot(times,bubble_meanADV)
        ax[1].plot(times,bubble_prop)
        
        init_vals_growA = [guess_grow[fs][ei],1,1]
        init_vals_grow = [guess_grow[fs][ei],1]
        
        try:
            best_vals_growA, covar_growA = curve_fit(bubble_growA, times, bubble_prop, p0=init_vals_growA)
            ax[1].plot(times,bubble_growA(times,*best_vals_growA))
        except:
            best_vals_grow, covar_grow = curve_fit(bubble_grow, times, bubble_prop, p0=init_vals_grow)
            ax[1].plot(times,bubble_grow(times,*best_vals_grow))
            
        fin_bubble.append([detuning[fs][ei], np.mean(bubble_mean[-2:-1]),np.mean(bubble_mean_err[-2:-1])])
        slope_bubble.append([detuning[fs][ei], np.mean(bubble_slopeADV[-3:-1]),np.mean(bubble_slopeADVerr[-3:-1])])
        grow_time.append([detuning[fs][ei], best_vals_growA[0],(covar_growA[0,0])**0.5])
        print(grow_time)
        half_time.append([detuning[fs][ei], best_vals_growA[0]*np.log((0.5+1)/best_vals_growA[2]),(covar_grow[0,0])**0.5])
        

        ax[2].plot(times,bubble_slope)
        ax[2].plot(times,bubble_slopeADV)
        ax[3].plot(np.sum(D,axis=1))        
        plt.show()
        
        plt.hist(bsize[np.where(bsize>0)],10)
        plt.xlim(0,200)


        #print(bubble_prop)
        #print(best_vals_growA[0],(covar_growA[0,0])**0.5)
        #print(best_vals_growA[1],(covar_growA[0,0])**0.5)
        #print(best_vals_growA[2],(covar_growA[0,0])**0.5)
        print(best_vals_growA[0]*np.log((0.5+1)/best_vals_growA[1]))
```
